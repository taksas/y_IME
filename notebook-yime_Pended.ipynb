```python
# import libraries
import time
from PIL import Image
from selenium import webdriver
from selenium.webdriver.common.by import By

import spacy
import ginza
import regex
import pykakasi

import glob
import re

import datetime


nlp = spacy.load('ja_ginza')
p1 = regex.compile(r'\p{Script=Han}+')
p2 = regex.compile(r'\p{Script=Latin}+')
p3 = regex.compile(r'\p{Numeric_Type=Numeric}+')
p4 = regex.compile('[0-9]+')
p5 = regex.compile('[０-９]+')
kks = pykakasi.kakasi()

dt_now = datetime.datetime.now()
realtime = dt_now.strftime('%Y%m%d%H%M%S')

def getter(url, subject):
    driver.get(url)
    elements = driver.find_elements_by_class_name('newsFeed_item_link')
    links = [i.get_attribute('href') for i in elements]
    for list in links:
        driver.get(list)
        element = driver.find_element_by_class_name('sc-kREsUy.fOkZWA')
        link2 = element.get_attribute('href')
        if 'articles' in link2:
            driver.get(link2)
            print(link2)
            text = driver.find_element_by_id('uamods').get_attribute("textContent")
            time = driver.find_element_by_class_name('sc-iGPElx.bEhhug').get_attribute("textContent")
            
            
            texta = text
            doc = nlp(texta)
  
            # 文節分割
            for sent in doc.sents:
                for span in ginza.bunsetu_spans(sent):
                    sp =  str(span)
        
                    if p1.search(sp) != None and p2.search(sp) == None and p3.search(sp) == None and p4.search(sp) == None and p5.search(sp) == None: 

                        yomi = ''
                        result = kks.convert(sp)
                        for result2 in result:
                            yomi += result2['hira']
            
          
                        try:
                            comment = "【" + time + " - " + subject + "】 " + driver.title.replace(" - Yahoo!ニュース", "")
                            f = open("dicpart" + realtime + '.txt', 'a', encoding="shift-jis")
                            output = yomi + '	' + sp.replace('\n', '')  + '	' + '名詞' + '	' + comment + '\n'
                            f.write(output.replace('\u3094', ''))
                        except Exception:
                            pass
            
            
            
            
            
            
            


chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument("--window-size=1920,1080")
driver = webdriver.Chrome(options=chrome_options)



getter('https://news.yahoo.co.jp/topics/domestic', "国内")
getter('https://news.yahoo.co.jp/topics/world', "国際")
getter('https://news.yahoo.co.jp/topics/business', "経済")
getter('https://news.yahoo.co.jp/topics/entertainment', "エンタメ")
getter('https://news.yahoo.co.jp/topics/sports', "スポーツ")
getter('https://news.yahoo.co.jp/topics/it', "IT")
getter('https://news.yahoo.co.jp/topics/science', "科学")
getter('https://news.yahoo.co.jp/topics/local', "地域")
print("finish")
    
```

```python
with open("output.txt", "w", encoding="shift-jis") as new_file:
    for name in glob.glob('./dicpart*.txt'):
        with open(name) as f:
            for line in f:
                new_file.write(line)
            
            new_file.write("\n")
```
